# JOIN

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
```

```python
df1
```

```python
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
df2
```

```python
df1.join(df2, lsuffix='_X', rsuffix='_Y')
```

```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

```python
df1.set_index('key')
```

```python
df2.set_index('key')
```

```python
df1.set_index('key').join(df2.set_index('key'))
```

```python
df1.join(df2.set_index('key'), on='key')
```

# Non-Unique key Values

```python
df3 = pd.DataFrame({'key': ['K0', 'K1', 'K1', 'K3', 'K0', 'K1'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
```

```python
df3
```

```python
df3.join(df2.set_index('key'), on='key')
```

```python
df1 = pd.DataFrame({'A':['K0','K1','K4','K7'],
'B':[45,23,45,2]})
df1
```

```python
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
'A': ['1', '2', '4', '23', '2', '78'],
'B': ['4', '41', '32', '23', '74', '5']})
df2
```

```python
df1.set_index('A').join(df2.set_index('key'),lsuffix='_caller', rsuffix='_other')
```

```python
df2.set_index('key').join(df1.set_index('A'),lsuffix='_caller', rsuffix='_other')
```

# (left join, Right join) 

```python
df1 = pd.DataFrame({'DF1_KEY':['K0','K1','K4','K7','K9'],
'B':[45,23,45,56,5]})
df1
```

```python
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
'A': ['1', '2', '4', '23', '2', '78'],
'B': ['4', '41', '32', '23', '74', '5']})
```

```python
df1.join(df2,how='left', lsuffix='_caller', rsuffix='_other')
```

```python
df1.join(df2,how='right', lsuffix='_caller', rsuffix='_other')
```
